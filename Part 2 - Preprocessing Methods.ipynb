## Blurring
### Box Blur

The Box Blur is a blurring algorithm that can, with enough iterations, implement a gaussian blur. The algorithm is fairly simple:

1) Center a "box" on a pixel.
2) Take samples of the values of all pixels within that box.
3) Create an average of those values.
4) Assign that average as the value of the corresponding pixel of the new image.
5) Repeat for a new pixel until exhausted.

A helpful reference on blurring can be located here:
https://mmeysenburg.github.io/image-processing/05-blurring/. The gif in particular is great.

````python
def box_blur(image, shape = [28,28], box_width = 3, iterations = 1):
    #Take a sample from image. For example, a 3*3 box/window of the original, as below.
    #  X X X   > upperRow
    #  X P X   > centerRow
    #  X X X   > bottomRow
    # Our goal value is the mean of all values in the window.
    # Since there are 9 values, sum them all and then divide by 9. i.e., the box_width**2.
    
    #But wait! What if our window goes outside the array, either below it or above it?
    #In that case, we need to take the maximum of (0, small edge) and the minimum of (len(array), large edge)
    #With small edge and large edge referring to small and large in the slice, array[small:large]
    #Our new sample window might look like this, or like this:
    # X X > upperRow                            #X X X > upperRow
    # P X > centerRow                           # X P X > centerRow
    # X X > bottomRow
    
    #Note: his means that the count of the values won't be box_width**2.
    #If we were writing this in c, we would have to find the count by hand.
    #But instead we can just use np.mean(our_window) to get the average of the values.

    half_width = int(box_width/2); #get half of the box_width for use in slicing.
    image = np.reshape(image, shape) #reshape the image into a 2-d array of the right shape.
    while 1: #repeat for the # of iterations desired.
        if iterations <= 0: break
        iterations -= 1; new_image = np.empty(shape)
        #decrement the count-down to prevent an infinite loop.
        #make an empty array to store the new pixel values in.
        for x,y in ((x, y) for y in range(shape[0]) for x in range(shape[1])):
            new_image[y,x] = min(int(np.mean(image[max((y - half_width), 0):min((y + half_width), shape[0]),
                                                   max((x - half_width), 0):min((x + half_width), shape[1])])), 255)
            #We don't want the window we're making to go above the length of the array, as that triggers an error.
            #We don't want the window we're making to go below 0, because then it encompasses the other side.
            #Thus, we want the max of either 0 or the left/topmost edge of the window (which might be negative)
            #Thus, we want the minimum of either the length of the array or the right/bottommost edge of the window (which might be too large)
            #Note that the new pixel needs to be an integer within the range 0-255.
            #Since all of the pixels are already 0+ and we gathered the average, there is no need to worry about pixels < 0.
        image = new_image #replace the old image with the new one.
    #If we're done making new images, return the one we have created.
    yield image.flatten()
````

#### Gaussian Blur
Note: The code that generates the kernel is from http://subsurfwiki.org/wiki/Gaussian_filter


Gaussian blur is more or less the same idea as box blur, except that rather than simply taking the average of all of the values within the box we are taking a *weighted* value of all of the values within the box.

How do we do this? We apply what is called a "kernel," which is defined thus: "In image processing, a kernel, convolution matrix, or mask is a small matrix. It is used for blurring, sharpening, embossing, edge detection, and more. This is accomplished by doing a convolution between a kernel and an image."

The gaussian kernel is a kernel which uses a gaussian distribution - think the bell curve that's taught in introductory statistics. More specifically, when used for a 2-d array it is generated according to the formula: G(x,y)= (2 pi sigma^2)^-1 e^(-(x^2 + y^2)/2 sigma ^2)). This kernel will be applied after we've selected the values we care about, a process described in the the box blur description.

Note that the gaussian kernel can be applied as separate 1-d transformations leading to the same end result. This process shifts the sampling cost from O N^2 to O N, making it much more efficient. Because the images we're dealing with are small (28 * 28) this isn't an issue and thus is not implemented.

Notably, because the box we're using to select may be shortened - or chopped off - , we also need to chop off the corresponding segments of the kernel. This can be done two ways: 

First, we can generate a kernel and then select from it using the exact same parameters used to select from the image: 

```image[max((y - half_width), 0):min((y + half_width), shape[0]),
      max((x - half_width), 0):min((x + half_width), shape[1])])```

Second, we can generate a smaller, box-sized kernel and subsequently select from it, a process described below:

We need to chop our kernel.
There are two situations in which the kernel needs to be chopped:
When the window is too low (y - half_width < 0)
When the window is too high (y + half_width > shape[0/1])

In the former case, for kernel [bottom:top] the bottom has been chopped off. The normal bottom of 0 needs to increase by the amount it has been chopped off. This amount is the difference between 0 and (y-half_width). I.e., the bottom will be determined by ```0 + max(0, -(y-half_width))```. The max will return 0 if y-half_width was positive, indicating nothing got chopped off, because the positive is made negative. The max will return > 0 if y-half_width was negative, indicating something got chopped off, because the negative was made positive.

In the latter case, for kernel[bottom:top] the top has been chopped off. The normal top of box_width+1 needs to decrease by the amount it has chopped off. This amount is the sum of the size of the array in the respective axis (size[0] or size[1]) - (y + half_width). I.e., the top will be determined by ```box_width+1+min(0,(size[axis] - (y+half_width)))```. If y + the half width > size, it will become negative, decreasing box_width. Otherwise, box_width will be unchanged due to the min function outputting 0.

We then multiply our box by our kernel and voila, we have the values we need to average.

````python
def gaussian_kernel(size, size_y=None): #code that generates a kernel, source described at bottom of above chunk.
    size = int(size)
    if not size_y:
        size_y = size
    else:
        size_y = int(size_y)
    x, y = np.mgrid[-size:size+1, -size_y:size_y+1]
    g = np.exp(-(x**2/float(size)+y**2/float(size_y)))
    return g / g.sum()

def gauss_blur(image, shape = [28,28], box_width = 3, iterations = 1):
    half_width = int(box_width/2);
    image = np.reshape(image, shape) 
    while 1:
        if iterations <= 0: break
        iterations -= 1; new_image = np.empty(shape)
        
        for x,y in ((x, y) for y in range(shape[0]) 
                           for x in range(shape[1])):
            sample = image[max((y - half_width), 0):min((y + half_width), shape[0]),
                           max((x - half_width), 0):min((x + half_width), shape[1])]
            kernel = gaussian_kernel(box_width)
            kernel = kernel[max(0, -(y-half_width)):(box_width + 1 + min(0, (shape[0] - (y + half_width)))),
                            max(0, -(x-half_width)):(box_width + 1 + min(0, (shape[1] - (x + half_width))))]
        image = new_image
        
    yield image.flatten()
````

## Pixel Shift

````python
def pix_shift(image, shape = [28, 28], shift = 1):
    yield image
    image = np.reshape(image, shape)
    #Up = topmost row goes to bottom
    split = np.split(image,[1*shift], axis=0)
    yield (np.concatenate((split[1],split[0]), axis=0)).flatten()
    
    #Down = bottomost row goes to top
    split = np.split(image,[-1*shift], axis=0)
    yield (np.concatenate((split[1],split[0]), axis=0)).flatten()
    
    #Left = leftmost row goes to right
    split = np.split(image, [1*shift], axis=1)
    yield (np.concatenate((split[1],split[0]), axis=1)).flatten()
    
    #Right = rightmost row goes to left
    split = np.split(image, [-1*shift], axis=1)
    yield (np.concatenate((split[1],split[0]), axis=1)).flatten()
````

## Noise Removal

Information on denoise_nl_means can be found at: http://scikit-image.org/docs/dev/api/skimage.restoration.html#skimage.restoration.denoise_nl_means. It uses the non-local means algorithm (citations: http://scikit-image.org/docs/dev/api/skimage.restoration.html#r833835). This abstract for the original paper - behind a paywall @ http://epubs.siam.org/doi/pdf/10.1137/040616024 (or through a link from the scikit-image page) - is as follows.

The search for efficient image denoising methods is still a valid challenge at the crossing of functional analysis and statistics. In spite of the sophistication of the recently proposed methods, most algorithms have not yet attained a desirable level of applicability. All show an outstanding performance when the image model corresponds to the algorithm assumptions but fail in general and create artifacts or remove image fine structures. The main focus of this paper is, first, to define a general mathematical and experimental methodology to compare and classify classical image denoising algorithms and, second, to propose a nonlocal means (NL-means) algorithm addressing the preservation of structure in a digital image. The mathematical analysis is based on the analysis of the "method noise," defined as the difference between a digital image and its denoised version. The NL-means algorithm is proven to be asymptotically optimal under a generic statistical image model. The denoising performance of all considered methods are compared in four ways; mathematical: asymptotic order of magnitude of the method noise under regularity assumptions; perceptual-mathematical: the algorithms artifacts and their explanation as a violation of the image model; quantitative experimental: by tables of L2 distances of the denoised version to the original image. The most powerful evaluation method seems, however, to be the visualization of the method noise on natural images. The more this method noise looks like a real white noise, the better the method.

````python
from scipy.ndimage import interpolation
def moments(image):
    c0,c1 = np.mgrid[:image.shape[0],:image.shape[1]] # A trick in numPy to create a mesh grid
    totalImage = np.sum(image) #sum of pixels
    m0 = np.sum(c0*image)/totalImage #mu_x
    m1 = np.sum(c1*image)/totalImage #mu_y
    m00 = np.sum((c0-m0)**2*image)/totalImage #var(x)
    m11 = np.sum((c1-m1)**2*image)/totalImage #var(y)
    m01 = np.sum((c0-m0)*(c1-m1)*image)/totalImage #covariance(x,y)
    mu_vector = np.array([m0,m1]) # Notice that these are \mu_x, \mu_y respectively
    covariance_matrix = np.array([[m00,m01],[m01,m11]]) # Do you see a similarity between the covariance matrix
    return mu_vector, covariance_matrix
def deskew(image):
    c,v = moments(image)
    alpha = v[0,1]/v[0,0]
    affine = np.array([[1,0],[alpha,1]])
    ocenter = np.array(image.shape)/2.0
    offset = c-np.dot(affine,ocenter)
    return interpolation.affine_transform(image,affine,offset=offset)
````
